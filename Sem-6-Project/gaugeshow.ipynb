```python
from datetime import date
from pandas_datareader import data as pdr
import numpy as np
import pandas as pd


import yfinance as yf

#Data viz
import plotly.graph_objs as go
import plotly as pt

import pandas_ta as ta
import matplotlib.pyplot as plt
from datetime import date
import pandas_datareader.data as webd
```

```python
import time
import threading






def get_data():
      data = yf.download(tickers='^NSEI', period='1d', interval='1m')
      gauge = go.Figure(go.Indicator(
        domain={'x': [0, 1],
                'y': [0, 1]},
        value=int(data.iloc[-1]['Open']),
        mode="gauge+number+delta",
        title={'text': "<b>NSEI DAY RANGE</b><br><span style='color: gray; font-size:0.8em'>Indian Rs</span>",
               'font': {"size": 20}},
        delta={'reference': data.iloc[-1]['Low']},
        gauge={
            'axis': {'range': [data.iloc[-1]['High'], data.iloc[-1]['Close']]},
            'steps': [
                {'range': [17250, 17750], 'color': "gray"}],
            'threshold': {'line': {'color': "red", 'width': 4},
                          'thickness': 0.75,
                          'value': int(data.iloc[-1]['Open'])}}))

      # gauge.show()
      pt.offline.plot(gauge, filename='./templates/mygauge.html', auto_open=False)



def printit():
  while True:
    get_data()
    time.sleep(20)


printit()
```
